### Design a multi-line plot using Matplotlib or Seaborn that shows the evolution of average mpg over the years, separately for each origin

```python
#your code here
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
Data = pd.read_csv(r'C:\users\samri\downloads\cars.csv')
Data= Data.dropna(subset=['mpg', 'origin', 'model_year'])



avg_mpg_by_year_origin = Data.groupby(['model_year', 'origin'])['mpg'].mean().reset_index()


plt.figure(figsize=(10, 6))
sns.lineplot(data=avg_mpg_by_year_origin, x='model_year', y='mpg', hue='origin', marker='o')


plt.title('Evolution of Average MPG Over Years by Origin')
plt.xlabel('Model Year')
plt.ylabel('Average MPG')
plt.xticks(range(Data['model_year'].min(), Data['model_year'].max() + 1))  
plt.legend(title='Origin')
plt.grid(True)
plt.tight_layout()

plt.show()
```

### Create a Seaborn scatterplot (or PairGrid) where:

X = horsepower

Y = weight

Color by: origin

Size by: mpg

Hue order = ['japan', 'europe', 'usa']

Add meaningful plot titles and axis titles.

```python
#your code here
Data = Data.dropna(subset=['horsepower', 'weight', 'origin', 'mpg'])


Data['origin'] = Data['origin'].str.lower()

hue_order = ['japan', 'europe', 'usa']

# Create the scatter plot
plt.figure(figsize=(10, 6))
scatter = sns.scatterplot(
    data=Data,
    x='horsepower',
    y='weight',
    hue='origin',
    size='mpg',
    sizes=(40, 400),
    hue_order=hue_order,
    palette='Set2',
    alpha=0.7
)


plt.title('Horsepower vs. Weight by Origin\n(Size represents MPG)', fontsize=14)
plt.xlabel('Horsepower', fontsize=12)
plt.ylabel('Weight', fontsize=12)
plt.legend(title='Origin', bbox_to_anchor=(1.05, 1), loc='upper left')  # Adjust legend position
plt.tight_layout()

plt.show()
```

### We define a consistent car model as one that was produced over multiple years and had very low variation in mpg across those years (standard deviation < 1.0).

Tasks:

Identify car names that appear in more than one model_year.

For each such name, compute the standard deviation of mpg across years.

Return the car(s) with the lowest variation in mpg, among those with at least 2 appearances and std(mpg) < 1.0.

Report the model name(s), number of appearances, and the average mpg.

Bonus:
Sort the result by number of appearances (descending), then mpg (descending).

```python
#your code here

Data = Data.dropna(subset=['mpg', 'model_year', 'name'])


grouped = Data.groupby('name')

# Step 2: Filter for cars appearing in multiple years
def appears_multiple_years(group):
    return group['model_year'].nunique() > 1

multi_year_models = grouped.filter(appears_multiple_years)

# Step 3: Group again and compute stats
result = (
    multi_year_models
    .groupby('name')
    .agg(
        num_years=('model_year', 'nunique'),
        appearances=('mpg', 'count'),
        mpg_std=('mpg', 'std'),
        avg_mpg=('mpg', 'mean')
    )
    .reset_index()
)


consistent = result[(result['mpg_std'] < 1.0) & (result['appearances'] >= 2)]


consistent_sorted = consistent.sort_values(by=['appearances', 'avg_mpg'], ascending=[False, False])


print(consistent_sorted[['name', 'appearances', 'avg_mpg']])
```
